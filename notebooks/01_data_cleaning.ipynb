```python
df['Selling Price'] = pd.to_numeric(df['Selling Price'].astype(str).str.replace(',','',regex=False),errors='coerce')
```

```python
df['Original Price'] = pd.to_numeric(df['Original Price'].astype(str).str.replace(',','',regex=False),errors='coerce')
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'].astype(str).str.replace(',','',regex=False),errors='coerce')
```

```python
df[df.duplicated(keep=False)]
```

```python
df.to_csv('fitness_trackers_cleaned.csv', index=False)
```
